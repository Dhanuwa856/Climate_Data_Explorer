```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
dates = pd.date_range(start='2024-01-01',periods=366, freq='D')
```

```python
np.random.rand(3)
```

```python
np.random.seed(100)
print(np.random.rand(3))
np.random.seed(100)  # යළි seed සකස් කිරීම

np.random.rand(3)
```

```python
temperature = np.random.normal(28,5, 366)
temperature 
```

```python
rainfall = np.random.poisson(lam=10,size=366)
rainfall
```

```python
wind_speed = np.random.uniform(5,25,366)
wind_speed
```

```python
climate_df = pd.DataFrame({
  'Date':dates,
  'Temperature':temperature,
  'Rainfall':rainfall,
  'Wind_speed':wind_speed
})
print(climate_df)
```

```python
climate_df.loc[10:15, 'Rainfall'] = np.nan
climate_df.loc[5:20,'Rainfall']
```

```python
climate_df["Rainfall"].fillna(climate_df["Rainfall"].mean(), inplace=True)
climate_df.loc[5:20,'Rainfall']
```

```python
climate_df['Temperature'] = np.round(climate_df['Temperature'],1)
climate_df['Temperature']
```

```python
climate_df['Month'] = climate_df['Date'].dt.month_name()
climate_df['Month']
```

```python
monthly_stats = climate_df.groupby('Month').agg(
  Avg_Temp = ('Temperature', 'mean'),
  Total_Rainfall = ('Rainfall', 'sum')
).reset_index()
monthly_stats
```

```python
plt.figure(figsize=(12,6))
plt.plot(climate_df['Date'],climate_df['Temperature'], label='Daily temperature', color='red' , alpha= 0.5)
plt.plot(climate_df['Date'], climate_df['Temperature'].rolling(7).mean(), color='black',label='Weekly average')
plt.title('2024 annual temperature trend',fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Temperature (°C)', fontsize=14)
plt.legend()
plt.grid(linestyle='--', alpha=0.5)
# plt.savefig("../result/temperature_trend.png")

plt.show()


```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

monthly_stats['Month'] = pd.Categorical(
    monthly_stats['Month'], 
    categories=month_order, 
    ordered=True
)

# මාස පිළිවෙලට සෝට් කරන්න
monthly_stats = monthly_stats.sort_values('Month')


plt.figure(figsize=(12,6))

plt.bar(monthly_stats['Month'], monthly_stats['Total_Rainfall'], color='blue', alpha=0.7, edgecolor='black')
plt.title('Monthly Rainfall in 2024', fontsize=16)
plt.xticks(rotation=45)
plt.ylabel('Total Rainfall (mm)', fontsize=14)
# plt.savefig('../result/rainfall_distribution.png')
plt.show()
```

```python
plt.scatter(climate_df['Temperature'], climate_df['Wind_speed'], alpha=0.7,  c=climate_df['Rainfall'], cmap='viridis')
plt.colorbar(label='Rainfall (mm)')
plt.title('Wind Speed VS Temperature')
plt.ylabel('Wind Speed (km/h)')
plt.xlabel('Temperature (°C)')
plt.savefig("../result/wind_speed_scatter.png")
plt.close()
```
